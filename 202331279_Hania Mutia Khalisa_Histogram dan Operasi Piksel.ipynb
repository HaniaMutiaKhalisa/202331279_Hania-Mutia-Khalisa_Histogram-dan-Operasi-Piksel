```python
##202331279_Hania mutia Khalisa
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

## membaca data gambar

```python
##202331279_Hania Mutia K
img= cv2.imread('mobil.jpg')
```

```python
img.shape
```

```python
(baris, kolom) = img.shape[:2]
```

## convert BGR to RGB

```python
##202331279_Hania mutia Khalisa
rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

## menampilkan histogram

```python
##202331279_Hania Mutia khalisa

fig, axs = plt.subplots(2,2,figsize=(15,5))

#cara1
axs[0,0].imshow(rgb)
axs[0,1].hist(rgb.ravel(),256,[0,256])

#cara2
hist = cv2.calcHist([rgb],[0],None,[256],[0,256])
axs[1,0].imshow(rgb)
axs[1,1].plot(hist)

#202331279_Hania Mutia KKhalisa
# menampilkan
plt.show()
```

## operasi piksel untuk meningkatkan kecerahan

```python
##202331279_Hania Mutia Khalisa
beta = 40
```

```python
citra_cerah = np.zeros((baris, kolom, 3))
```

```python
##202331279_Hania Mutia Khalisa
for x in range(baris):
    for y in range(kolom):
        gyx = rgb[x,y] + beta
        citra_cerah[x,y] = gyx
```

```python
citra_cerah = citra_cerah.astype(np.uint8)
```

```python
##202331279_Hania Mutia K
fig, axs = plt.subplots(2, 2,figsize=(15,5))
axs[0,0].imshow(rgb)
axs[0,1].hist(rgb.ravel(),256,[0,256])
axs[1,0].imshow(citra_cerah)
axs[1,1].hist(citra_cerah.ravel(),256,[0,256])

plt.show()
```

## meregangkan kontras

```python
##202331279_Hania Mutia K
alpha = 1.2
```

```python
citra_kontras = np.zeros_like(rgb, dtype=np.float32)
```

```python
##202331279_Hania Mutia K
for x in range(baris):
    for y in range(kolom):
        gyx = alpha * rgb[x, y]
        citra_kontras[x,y] = np.clip(gyx, 0, 255)
```

```python
citra_kontras = citra_kontras.astype(np.uint8)
```

```python
##202331279_Hania Mutia k
fig, axs = plt.subplots(2, 2,figsize=(15,5))
axs[0,0].imshow(rgb)
axs[0,0].set_title("Gambar asli")
axs[0,1].hist(rgb.ravel(),256,[0,256])
axs[1,0].imshow(citra_kontras)
axs[0,0].set_title("gambar setelah kontras")
axs[1,1].hist(citra_kontras.ravel(),256,[0,256])

plt.show()
```

## operasi gabungan antara meningkatkan kecerahan dan meregangkan kontras

```python
##202331279_Hania Mutia K
beta = 10
alpha = 1.2
```

```python
baris, kolom, _ = rgb.shape
```

```python
citra_kontras = np.zeros_like(rgb, dtype=np.float32)
```

```python
## 202331279_Hania Mutia K
for x in range(baris):
    for y in range(kolom):
        for c in range(3):
            nilai = alpha * rgb[x, y, c] + beta
            citra_kontras[x, y, c] = np.clip(nilai, 0, 255)
```

```python
## 202331279_Hania Mutia K
citra_kontras = citra_kontras.astype(np.uint8)

fig, axs = plt.subplots(2,2,figsize=(15,5))
axs[0,0].imshow(rgb)
axs[0,0].set_title('gambar asli'), 
axs[0,1].hist(rgb.ravel(), bins=256)
axs[0,1].set_title('histogram asli')
axs[1,0].imshow(citra_kontras)
axs[1,1].hist(citra_kontras.ravel(),256,[0,256])
axs[1,1].set_title('histogram setelah operasi')

plt.tight_layout()
plt.show()
```

## membalik citra

```python
#202331279-Hania mutia k
citra_negatif = np.zeros((baris, kolom, 3), dtype=np.uint8)
```

```python
#202331279_Hania Mutia k
for x in range(baris):
    for y in range(kolom):
        for c in range(3): 
            citra_negatif[x, y, c] = 255 - rgb[x, y, c]
```

```python
citra_negatif = np.zeros((baris, kolom, 3), dtype=np.uint8)
```

```python
#202331279_Hania Mutia k
fig, axs = plt.subplots(2,2,figsize=(15,5))
axs[0,0].imshow(rgb)
axs[0,0].set_title('gambar asli')
axs[0,1].hist(rgb.ravel(), bins=256, range=[0,256])
axs[0,1].set_title('histogram asli')

axs[1,0].imshow(citra_negatif)
axs[1,1].set_title('gambar negatif')
axs[1,1].hist(citra_negatif.ravel(), bins=256, range=[0,256])
axs[1,1].set_title('histogram negatif')

plt.tight_layout()
plt.show()
```
